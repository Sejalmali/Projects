```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def add_gaussian_noise(image, mean=0, sigma=25):
    """Add Gaussian noise to an image."""
    gauss = np.random.normal(mean, sigma, image.shape)
    noisy_image = np.clip(image + gauss, 0, 255)
    return noisy_image.astype(np.uint8)
```

```python
def apply_average_filter(image, kernel_size=3):
    """Apply average filter to an image."""
    return cv2.blur(image, (kernel_size, kernel_size))
```

```python
def apply_median_filter(image, kernel_size=3):
    """Apply median filter to an image."""
    return cv2.medianBlur(image, kernel_size)
```

```python
image_path = r"/content/IMG_20240109_000528.jpg"
image = cv2.imread(image_path)
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')
```

```python
noisy_image = add_gaussian_noise(image)
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 2)
plt.imshow(cv2.cvtColor(noisy_image, cv2.COLOR_BGR2RGB))
plt.title("Noisy Image")
plt.axis('off')
```

```python
average_filtered_image = apply_average_filter(noisy_image)
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 3)
plt.imshow(cv2.cvtColor(average_filtered_image, cv2.COLOR_BGR2RGB))
plt.title("Average Filtered Image")
plt.axis('off')
```

```python
median_filtered_image = apply_median_filter(noisy_image)
```

```python
plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 4)
plt.imshow(cv2.cvtColor(median_filtered_image, cv2.COLOR_BGR2RGB))
plt.title("Median Filtered Image")
plt.axis('off')
```
